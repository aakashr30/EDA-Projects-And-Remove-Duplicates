```python
p100 = data.Purchase.max()
p0 = data.Purchase.min()

q1 = data.Purchase.quantile(0.25)
q2 = data.Purchase.quantile(0.50)
q3 = data.Purchase.quantile(0.75)

iqr = q3-q1  #inter quantile range
```

```python
lower_cutoff = q1 - 1.5*iqr

upper_cutoff = q3 + 1.5*iqr
```

```python
lower_cutoff,upper_cutoff
```

```python
 # IF lower_cutoff<p0 there are no outliers o the lower side
 #if upper_cutoff >p100 there are no outliers on the higher side
```

```python
print("p0 ",p0,
      "p100 =",p100,
      "lc =  ",lower_cutoff,
      "uc = ",upper_cutoff)
```

```python
data.Purchase.clip(upper = upper_cutoff)
plt.boxplot(data['Purchase'])
```

```python
#from IPython.core.pylabtools import figsize
data.Purchase.clip(upper = upper_cutoff,inplace = True)
fig,ax=plt.subplots(figsize=(12,10))
plt.boxplot(data['Purchase'])
```
